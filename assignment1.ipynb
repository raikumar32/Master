## 3.Data Cleaning

```python
#count the number of missing value in each column
nba_player.isnull().sum()
```

```python
#checking duplicate rows
duplicateRows = nba_player[nba_player.duplicated()]
duplicateRows
```

```python
nba_player['Pos'].value_counts()
```

```python
nba_player['Player'].value_counts()
```

```python
nba_player['Tm'].value_counts()
```

```python
# Error 1
# Removing tailing Whitespace from Pos column
nba_player['Pos']=(nba_player.loc[:,'Pos'].str.strip())
```

```python
# Removing tailing Whitespace from Tm column
nba_player['Tm']=(nba_player.loc[:,'Tm'].str.strip())
```

```python
# Error 2: Invalid values(typos) in column Pos
# Returning a Series containing counts of unique values of Pos
nba_player['Pos'].value_counts()
```

```python
# Assigning valid values of Pos in filt_pos
filt_pos=['PF','PG','C','SG','SF','PG-SG','SF-PF']
```

```python
# Fetching values from column Pos which are valid
nba_player.loc[(nba_player['Pos'].isin(filt_pos))]
```

```python
# Fetching values from column Pos which are not valid
nba_player.loc[~(nba_player['Pos'].isin(filt_pos))]
```

```python
# Replacing Invalid value with valid ones
# https://stackoverflow.com/questions/28706968/how-to-correct-typos-in-pandas-dataframe
nba_player['Pos'] = nba_player['Pos'].replace(['SGa','SF.','sg','Sf','pg','Pg','PFa'],['SG','SF','SG','SF','PG','PG','PF'])
```

```python
# Returning a Series containing counts of unique values of Pos
nba_player['Pos'].value_counts()
```

```python
# Error 3: Impossible values in column Age
# Returning a Series containing counts of unique values of Age
nba_player['Age'].value_counts()
```

```python
# Replacing Invalid age "-19" with valid one
nba_player['Age']=nba_player['Age'].replace([-19],[19])
```

```python
# Returning a Series containing counts of unique values of Age
nba_player['Age'].value_counts()
```

```python
# Getting the Rows whose palyer's age is 280
nba_player.loc[nba_player['Age']==280]
```

```python
# Creating nan using numpy
nba_player.loc[nba_player['Player']=='Anthony Gill','Age']=np.nan
# Displaying nan value
nba_player[nba_player['Player']=='Anthony Gill'].head()
```

```python
# Replacing missing data with mean value
nba_player['Age'].fillna(nba_player['Age'].mean(axis=0),inplace=True)
# Rounding age column 
# https://stackoverflow.com/questions/26133538/round-a-single-column-in-pandas
nba_player['Age'] = nba_player['Age'].round()
```

```python
# Returning a Series containing counts of unique values of Age
nba_player['Age'].value_counts()
```

```python
# Error 4: correcting typos in column Tm
# Returning a Series containing counts of unique values of Team
nba_player['Tm'].value_counts()
```

```python
# Creating variable with valid values of Teams and assigning it to "filt_tm"
filt_tm=['MIA','MIL','NOP','SAS','PHO','MEM','TOT','BRK','CLE','ORL','LAL','POR','TOR','CHI','WAS','UTA','SAC','CHO','NYK','DEN','LAC','GSW','OKC','MIN','DET','DAL','IND','ATL','PHI','BOS','HOU']
```

```python
# Fetching values from column Tm which are valid
valid_tm = (nba_player['Tm'].isin(filt_tm))
nba_player[valid_tm]
```

```python
# Fetching values from column Tm which are not valid
# https://stackoverflow.com/questions/43256402/pandas-analogue-of-sqls-not-in-operator
nba_player[~(nba_player['Tm'].isin(filt_tm))]
```

```python
# Replacing invalid values in Tm column with valid ones
# https://stackoverflow.com/questions/28706968/how-to-correct-typos-in-pandas-dataframe
nba_player['Tm'] = nba_player['Tm'].replace(['H0U','NyK'],['HOU','NYK'])
```

```python
# Returning a Series containing counts of unique values of Team
nba_player['Tm'].value_counts
```

```python
# Error 5
# Printing columns having NaN values 
# https://stackoverflow.com/questions/37366717/pandas-print-column-name-with-missing-values
nba_player[nba_player.columns[nba_player.isnull().any()]]
```

```python
# Replacing NaN value with 0 for column FG%,3P%,2P% and FT%
nba_player['FG%'].fillna(0,inplace=True)
nba_player['3P%'].fillna(0,inplace=True)
nba_player['2P%'].fillna(0,inplace=True)
nba_player['FT%'].fillna(0,inplace=True)
```

```python
# Fetching rows from nba_player whose PTS value is greater than 1999
nba_player.loc[nba_player['PTS']>1999]
```

```python
# Error 6: Value of PTS is greater than 1999 for Rk 2 and 5
# Calculating total points
total_points=((nba_player['2P']*2)+(nba_player['3P']*3)+nba_player['FT'])
# Assigning calulated total points to PTS column in the data frame 
nba_player['PTS']=total_points
# Displaying head of nba_player dataframe
nba_player.head()
```

## 4.Save the Output

```python
# Writing Clean .csv file 'cleaned_NBA_players_stats.csv'
nba_player.to_csv('cleaned_NBA_players_stats.csv')
```

# Task 2: Data Exploration


## Task 2.1 
Explore the players' total points: Please analyze the composition of the total points of the top five players with the most points.

```python
# Libraries
from matplotlib import pyplot as plt
```

```python
# Showing descriptive analytics of nba_player
nba_player.describe()
```

```python
# Group by players and sort by total points in descending order
#https://stackoverflow.com/questions/27842613/pandas-groupby-sort-within-groups
top5players_pts=(nba_player.groupby(['Player'])['PTS'].sum().sort_values(ascending=False).head(5))
print(top5players_pts)
explode = [0.1, 0, 0, 0, 0]
```

```python
# Plot pie chart showing comosition of top five players with most points
plt.pie(top5players_pts,labels=top5players_pts.index,explode=explode,shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})
plt.title("Top Five Players with most points")
plt.tight_layout()
plt.show()
```
